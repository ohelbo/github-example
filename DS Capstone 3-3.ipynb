```python
# Initialising
from bs4 import BeautifulSoup
from tabulate import tabulate
import numpy
import requests


import pandas as pd
print("OK")
```

```python
# Screen scrape and putting table into dataframe
res = requests.get("http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))[0]

#Create Dataframe with postcodes and remove/update obsolete records
postcodes_list =  list(zip(df[0],df[1], df[2]))
df_postcod1 = pd.DataFrame(postcodes_list, columns = ['Postcode','Borough', 'Nighbourhood'])

df_postcod1.drop(0,inplace=True)
df_postcod1['Borough'] = df_postcod1['Borough'].astype('str')
df_postcod2 = df_postcod1[df_postcod1.Borough != 'Not assigned']
df_postcod2.loc[df_postcod2.Nighbourhood == 'Not assigned','Nighbourhood'] = df_postcod2.Borough

#Merge Nighbourhoods for postcodes
postcodes = df_postcod2.Postcode.unique()
for postcode in postcodes:
    postcode_subset = df_postcod2.loc[(df_postcod2.Postcode == postcode)]
    post_nigh = ', '.join(postcode_subset.Nighbourhood.unique())
    df_postcod2.loc[(df_postcod2.Postcode == postcode),'Nighbourhood'] = post_nigh

#Unique postcodes
df_postcod2.drop_duplicates(subset ="Postcode", keep = 'first', inplace = True) 

df_postcod2
# As on 3-1
```

```python
#Initiate the geospacial data
import pandas as geo_pd 
geo_data = geo_pd.read_csv("http://cocl.us/Geospatial_data") 

geo_data.rename(columns={'Postal Code':'PostalCode'}, inplace = True)

print(geo_data.shape)
geo_data.head()
```

```python
#Adding Latitude and Longitude to df_postcod2
pos_post  = geo_data['PostalCode'].astype('str')
pos_lat = geo_data ['Latitude'].astype('float')
pos_long = geo_data ['Longitude'].astype('float')

rown=0
for item in pos_post:
    postcode = str(item)
    df_postcod2.loc[df_postcod2.Postcode == postcode,'Latitude'] = pos_lat[rown]
    df_postcod2.loc[df_postcod2.Postcode == postcode,'Longitude'] = pos_long[rown]
    rown = rown+1
    
print(df_postcod2.shape)
df_postcod2.head(20)
# As on 3-2
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, nighbourhood in zip(df_postcod2['Latitude'], df_postcod2['Longitude'], df_postcod2['Borough'], df_postcod2['Nighbourhood']):
    label = '{}, {}'.format(nighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
