```python
# Initialising
from bs4 import BeautifulSoup
from tabulate import tabulate
import numpy
import requests


import pandas as pd
print("OK")
```

```python
# Screen scrape and putting table into dataframe
res = requests.get("http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))[0]

#Create Dataframe with postcodes and remove/update obsolete records
postcodes_list =  list(zip(df[0],df[1], df[2]))
df_postcod1 = pd.DataFrame(postcodes_list, columns = ['Postcode','Borough', 'Nighbourhood'])

df_postcod1.drop(0,inplace=True)
df_postcod1['Borough'] = df_postcod1['Borough'].astype('str')
df_postcod2 = df_postcod1[df_postcod1.Borough != 'Not assigned']
df_postcod2.loc[df_postcod2.Nighbourhood == 'Not assigned','Nighbourhood'] = df_postcod2.Borough

#Merge Nighbourhoods for postcodes
postcodes = df_postcod2.Postcode.unique()
for postcode in postcodes:
    postcode_subset = df_postcod2.loc[(df_postcod2.Postcode == postcode)]
    post_nigh = ', '.join(postcode_subset.Nighbourhood.unique())
    df_postcod2.loc[(df_postcod2.Postcode == postcode),'Nighbourhood'] = post_nigh

#Unique postcodes
df_postcod2.drop_duplicates(subset ="Postcode", keep = 'first', inplace = True) 

print(df_postcod2.shape)
df_postcod2
```
